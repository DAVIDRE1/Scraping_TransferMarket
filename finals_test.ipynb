```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import time
from datetime import datetime
import numpy as np
```

```python
torneo = "liga-dimayor-clausura"
fase = "final"
```

```python
partidos = pd.DataFrame(columns=['year', 'torneo', "fase", 'jornada', 'local', 'goles_local', 'goles_visitante', 'visitante', 'fecha', 'dia', 'arbitro', 'pos_local', 'pos_visitante', 'gana_local', 'empate', 'gana_visitante', 'url'])
```

```python

def scrap_page(url, year, fase):
    # Create a WebDriver instance for Chrome
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
    driver.maximize_window()
    # Visit the website
    # driver.get(url)

    # # Click on accept cookies policy
    # time.sleep(2)
    # WebDriverWait(driver, 20).until(EC.frame_to_be_available_and_switch_to_it((By.ID, "sp_message_iframe_851961")))
    # WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[@aria-label='Aceptar']"))).click()
    # time.sleep(2)

    driver.get(url)
    div_element = driver.find_element(By.CSS_SELECTOR, 'div.box.sptg-print.swiper-container')
    boxes = div_element.find_elements(By.XPATH, './div/table/tbody/*')
    
    for i in range(len(boxes)):
        # try:
        #     fase = boxes[i].find_element(By.XPATH, f'''./td/h2''').text
        # except NoSuchElementException:
            l = [year, torneo, fase, np.nan]
            l.append(boxes[i].find_element(By.XPATH, f'''./td[4]/a''').get_attribute("innerText")) # home
            try:
                l.append((boxes[i].find_element(By.XPATH, f'''./td[6]/a''').get_attribute("innerText").split(":")[0])) # goals home
            except NoSuchElementException:
                l.append(np.nan) # goals home
            try:
                l.append((boxes[i].find_element(By.XPATH, f'''./td[6]/a ''').get_attribute("innerText").split(":")[1])) # goals visitor
            except NoSuchElementException:
                l.append(np.nan) # goals visitor
            l.append(boxes[i].find_element(By.XPATH, f'''./td[8]/a''').get_attribute("innerText")) # visitor
            try:
                l.append(datetime.strptime(boxes[i].find_element(By.XPATH, f'''./td[1]/a''').get_attribute("innerText"), "%d/%m/%Y").date()) # date
            except NoSuchElementException:
                l.append(np.nan) # date
            l.append(boxes[i].find_element(By.XPATH, f'''./td[1]''').get_attribute("innerText").split(" ")[0]) # day
            l.append(np.nan) # referee
            l.append(np.nan) # pos home
            l.append(np.nan) # pos visitor        
            if l[5] > l[6]:
                l=l+[True,False,False]
            elif l[5] == l[6]:
                l=l+[False,True,False]
            else:
                l=l+[False,False,True]
            l.append(boxes[i].find_element(By.XPATH, f'''./td[6]/a''').get_attribute("href")) # url
            partidos.loc[len(partidos)] = l
    driver.quit()
    return partidos, fase
```

```python
for i in range(2023,2024):
    print(i)
    website = f"https://www.transfermarkt.es/fase-final-liga-dimayor-clausura/startseite/wettbewerb/COLO/plus/?saison_id={i-1}"
    print(website)
    partidos, fase = scrap_page(website, i, fase)
```
